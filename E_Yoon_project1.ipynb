```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/jtopor/DAV-5400/master/Project1/hflights.csv")
```

We assume that we are planning to travel and want to reduce any risks that may affect to our schedule for the trip. Therefore, we want to avoid airines which have high cancellation and high level of delay. Also, we want to avoid the day of week where the rate of the cancellation is the highest. In that way, we could reduce the risk so that we can make our trip run smoothly. 

Question 1: Which carriers have high mean of arrival delay?

Answer: EV(EVA Airline) has the highest mean of arrival delay with 13.93 minutes and F9(Frontier Airline) has the second highest mean of arrival delay with 13.73 minutes. Therefore, we can conclude that it is better to avoid those two airlines if we can.

```python
groupbyUC = df.groupby(['UniqueCarrier']).mean() #Grouping by carrier and find the mean of arrival delay for each carriers
print(groupbyUC)
bars = groupbyUC.plot.bar(y='ArrDelay') #Make it a bar chart and show the result
bars;
```

Question 2: Which carriers have high number of cancellation? 

Answer: The top three airlines with high number of cancellation: CO(Continental Airlines) with 6830, XE(ExpressJet Airlines) with 6775, and WN(Southwest Airlines) with 3708. Therefore, we can conclude that we need to avoid those three airline as possible. 

```python
groupbyUC1 = df.groupby(['UniqueCarrier']).count() #Grouping by carriers and counting the number of cancellation. 
print(groupbyUC1)
bars = groupbyUC1.plot.bar(y='Cancelled') #Visualizing the result as a bar chart
bars;
```

Question 3: Which day of week has the highest cancellation rate?

Answer: The day where has the highest cancellation rate is Monday. Therefore, we have to avoid the flight on Monday in order to reduce the risk of cancellation. Instead, Saturday would be the best day to minimize the risk of cancellation since it has the lowest rate of cancellation. 

```python
flight2 = df[df['Cancelled']==1] #Subset where the cancellation = 1 from the dataflight and define it as flight2
print(flight2)
group_DayOfWeek = flight2.groupby('DayOfWeek').count() #Groupby the day of week for the subset flight2
print(group_DayOfWeek)
group_DayOfWeek.plot(kind='bar', y='Cancelled'); #Make a bar chart of cancellation for each day of week 
```
