## 1. Introduction

TED Talks are influential videos from expert speakers on education, business, science, technology, entertainment and design etc. with subtitles in more than 100 languages. On their website, they are building a clearinghouse of free knowledge from the world's most inspired thinkers and we would love to be part of them. ## 2. Data Summary 

We got a dataset from Kaggle (https://www.kaggle.com/rounakbanik/ted-talks). There are total 2,550 use cases in the data set and 17 attributes. Details of the attributes are below:

* name: The official name of the TED Talk. Includes the title and the speaker.
* title: The title of the talk
* description: A blurb of what the talk is about.
* main_speaker: The first named speaker of the talk.
* speaker_occupation: The occupation of the main speaker.
* num_speaker: The number of speakers in the talk.
* duration: The duration of the talk in seconds.
* event: The TED/TEDx event where the talk took place.
* film_date: The Unix timestamp of the filming.
* published_date: The Unix timestamp for the publication of the talk on TED.com
* comments: The number of first level comments made on the talk.
* tags: The themes associated with the talk.
* languages: The number of languages in which the talk is available.
* ratings: A stringified dictionary of the various ratings given to the talk (inspiring, fascinating, jaw dropping, etc.)
* related_talks: A list of dictionaries of recommended talks to watch next.
* url: The URL of the talk.
* views: The number of views on the talk.

Among those attributes, we will only use these 5 attributes and the types of each attributes are below:

1. duration - numerical variable
2. views - numerical variable
3. comments - numerical variable
4. published_date - categorical variable
5. speaker_occupation - categorical variable

## 3. Exploratory Data Analysis (EDA) 

* duration

```python
#The data of duration of the talk is in seconds, so change it to minutes
df['duration'] = df['duration']/60 

df["duration"].describe() #summary statistics
```

```python
duration = df['duration']/60 # define duration in dataset as duration 

plt.figure(figsize=(10,5)) # define a new figure for plotting

plt.subplot(121) #subplot ('122' = 1x2, position 1)

plt.hist(duration) # define the type of plot as a histogram 

plt.title('Histogram of duration (Matplotlib)') # give the plot a title

plt.xlabel(' ') # blank out the x-axis label 

plt.subplot(122) #subplot ('122' = 1x2, position 2)

sns.distplot(duration, bins = 10) # specify the number of bins to use in the histogram

plt.title('Histogram of duration (Seaborn)') # give the plot a title

plt.show();
```

On average, Ted talks are about 13-14 minutes. So we better keep that in mind when we prepare for the talk so that audiences will not lose of focus of our talk.

* published date

```python
df['published_date'].count() # summary statistics
```

```python
#Using a unix epoch time
import datetime
df['published_date']=pd.to_datetime(df['film_date'],unit='s')

df['published_date']
```

```python
# count the number of each years
published_date_year = df.groupby(df['published_date'].dt.year)['published_date'].count()
published_date_year
```

```python
plt.figure(figsize=(10,5)) # Define a new figure and adjust figure size

plt.subplot(121) #subplot ('122' = 1x2, position 1)

sns.lineplot(data = published_date_year) # define the type of plot as a linechart 

plt.title('Line chart of published date (Seaborn)') # give a title

plt.xlabel('') # blank out the x-axis label

plt.subplot(122) #subplot ('122' = 1x2, position 2)

plt.plot(published_date_year, 'go--'); # define the type of plot as a linechart and change line type and color

plt.title('Line chart of published date (Matplotlib)') # give a title

plt.show();
```

After Ted talk was founded in 1984, its popularity has been increased over the years. The line graph clearly shows the trend its popularity and popularity increased dramatically since 2009. The values for 2017 is pretty low compare to the recent years. We believe that the data might be collected in the beginning of the year of 2017 therefore there's small amount of data for the year of 2017. 

* speaker's occupation

```python
df[].count() # summary statistics
```

```python
# grouped the occupation by the number of comments in order to sort out top 10
occupation = df.groupby(['speaker_occupation']).count()['comments'] 
occupation_df = occupation.sort_values(ascending=False) # sort out values
top10_occupation = occupation_df.head(10) # get the result
top10_occupation
```

```python
plt.figure(figsize=(32,10)) # make a new figure and adjust figure size
plt.subplot(121) #subplot ('122' = 1x2, position 1)
sns.barplot(x=top10_occupation.index,y=top10_occupation.values) #
plt.title("Bar chart of speaker's occupation (Seaborn)") #Create a title
plt.xlabel('')

plt.subplot(122) #subplot ('122' = 1x2, position 2)
top10_occupation.plot.bar();  # define the type of plot as a barchart 

plt.title("Bar chart of speaker's occupation (Matplotlib)") #Create a title

plt.show();
```

Since there are 1,458 different occupations for TED talk speakers, we decided to use only top 10 occupations to show our result. The bar charts above show that the top occupation for TED talk speakers is writer. Because other top occupations are artist, designer, journalist, we conclude that it would be better for us to have a creative job to increase a chance to get a popularity of our TED talks.  

## 4. Inference 

```python
df.corr() #correlation matrix for the data
```

```python
sns.jointplot(x='views', y='comments', data=df)
```

From the correlation matrix, we were able to find there are high correlation between views and comments than any other factors. Thus, we decided to draw the scatterplot for that as well. Since the coefficient is slightly more than 0.5, we can conclude that there is a medium to strong correlation between these two factors. In other words, if a video has more views it usually has more comments. Our goal is generating high engagement from audiences for both views and comments since they are the factors to measure the popularity.

```python
comcorr = df.corr()

k = 10 #number of variables for heatmap
f, ax = plt.subplots(figsize=(12, 9))
cols = comcorr.nlargest(k, 'views')['views'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 15}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

We wanted to make the correlation matrix visually attractive, so tried to draw a heatmap using  seaborn. In that way, it is now much easier to read the data. We learned that how powerful data visualization it is to make viewers to read the data easily and quickly.  

```python
#create a new column with month of date field 'published_date'
df['month'] = pd.DatetimeIndex(df['published_date']).month
df.head()
```

```python
# group the data from published_date by months 
# then define the number of talks for each month as month as month to draw the bar plot
month = (df.groupby('month').count())['published_date']
plt.figure(figsize=(15,6))
sns.barplot(x=month.index,y=month.values) #x-axis as month and y-axis as values
plt.show()
```

From the graph above, clearly February is the most popular month for TED talks while January and August are the least popular months. To get a higher chance of popularity, we aim to give a talk on Febuary and avoid January and August if possible.
