## 1. My Student Test Data

```python
# Load my .CSV file into my GitHub repository
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/yoonbs0123/DAV-5400/master/project3.untidy1.csv')
df
```

```python
# Reshaping data in Pandas to tidy and transform the data (wide to long format)
ts = pd.melt(
    df,
    id_vars=['Student '],
    var_name=['Subject'],
    value_name='Score')

print(ts)
```

The questions are what is the average overall scores for each person and the highest or lowest scores in certain subject. I grouped by students then calculated the mean to find average scores for each student. To find the highest and lowest scores for each subject, I grouped by subject then calculated the min and max values. Therefore, I can conclude that the average, lowest, highest socres for each student/subject are below.  

```python
#average overall scores for each person
avt = ts.groupby(['Student ']).mean()
print(avt)
```

```python
#the lowest scores for each subject
mint = ts.groupby(['Subject'])['Score'].min()
mint
```

```python
#the highest scores for each subject
maxt = ts.groupby(['Subject'])['Score'].max()
maxt
```

## 2. Natan's Test grades for students

```python
# Load CSV file into my GitHub repository
ds = pd.read_csv('https://raw.githubusercontent.com/yoonbs0123/DAV-5400/master/project3.untidy2.csv')
ds
```

```python
# Reshaping data in Pandas to tidy and transform the data (wide to long format)
test = pd.melt(
    ds,
    id_vars=['Name'],
    var_name=['Test number'],
    value_name='Score')

print(test)
```

The questions he asked were: 
##### 1. Which student had the highest average test grades?
I grouped by Name then found a mean of each. I can conclude that Jim Jackson has the highest average test grades among his classmates with the score 93


##### 2. Which student had the greatest range of test grades?
To find a range of test grades I grouped by student then calculate the value for max-min for each student. After I did that, I sorted in ascending order that highest values will be shown on the last. Now I can see that there are two students who had the greatest range of test grades: Chris Webber and Reggie Miller. I used [-2:] to print their name with values. 


##### 3. Which test had the highest grades?
To find the highest grade, I grouped by Test number then max of the score. Therefore, I can conclude that the highest grades for each test are 97.0 for Test1, 91.0 for Test2, 94.0 for Test3, and 95.0 for Test4. Among these four tests, Test1 has the highest value. 

```python
#Which student had the highest average test grades
avg_test = test.groupby('Name').mean() #group by name and find a mean
print(avg_test)
avg_test.max()
```

```python
#Which student had the greatest range of test grades
#grouped by student then calculate the value for max-min
gap = test.groupby('Name').max()['Score'] - test.groupby('Name').min()['Score'] 
gap.sort_values(inplace = True)
gap
```

```python
#print last 2 values
gap[-2:]
```

```python
#Which test had the highest grades
maxtt = test.groupby(['Test number'])['Score'].max() #group by test number then find the max of the score for each
maxtt
```

# 4. Avi's County College Grad/Jobs Data (Bonus)

*** I just did it as an extra and was hoping that I can practice more and try different functions for different data analysis ***

```python
# Load CSV file into my GitHub repository
dc = pd.read_csv('https://raw.githubusercontent.com/yoonbs0123/DAV-5400/master/project3.untidy4.csv')
dc
```

```python
# Converting the data table from wide format to long format
job = pd.melt(
    dc,
    id_vars=['Country'],
    var_name=['Year'],
    value_name='Number')

job
```

```python
#create a new dataset 'job_variables' for the LandArea and NatAmenity
job_variables = job[(job['Year']=='LandArea')|(job['Year']=='NatAmenity')]
job_variables
```

```python
#create a new dataset 'job_variables2' for the variables other than LandArea and NatAmenity, which are College/Jobs
job_variables2 = job[(job['Year']!='LandArea')&(job['Year']!='NatAmenity')]
job_variables2
```

One thing I learned from the process to do analysis from the long format, I was not able to generate visualized graph/plot for that. Therefore, I had to make them back to wide format by using pivot function. Then I was able to make a correlations chart for each.

```python
pivotjob = job_variables.pivot('Country', 'Year', 'Number')
pivotjob
```

```python
pivotjob2 = job_variables2.pivot('Country', 'Year', 'Number')
pivotjob2
```

```python
#Create Correlation pivotjob
corrjob = pivotjob.corr()

#Plot figsize
fig, ax = plt.subplots(figsize=(10, 10)) 

#Generate Color Map
colormap = sns.diverging_palette(220, 10, as_cmap=True)

#Generate Heat Map, allow annotations and place floats in map
sns.heatmap(corrjob, cmap=colormap, annot=True, fmt=".2f")

#Apply xticks
plt.xticks(range(len(corrjob.columns)), corrjob.columns);

#Apply yticks
plt.yticks(range(len(corrjob.columns)), corrjob.columns)

#show plot
plt.show()
```

```python
#Create Correlation pivotjob2
corrjob2 = pivotjob2.corr()

#Plot figsize
fig, ax = plt.subplots(figsize=(10, 10)) 

#Generate Color Map
colormap = sns.diverging_palette(220, 10, as_cmap=True)

#Generate Heat Map, allow annotations and place floats in map
sns.heatmap(corrjob2, cmap=colormap, annot=True, fmt=".2f")

#Apply xticks
plt.xticks(range(len(corrjob2.columns)), corrjob2.columns);

#Apply yticks
plt.yticks(range(len(corrjob2.columns)), corrjob2.columns)

#show plot
plt.show()
```

His questions are if LandArea or NaAmenity has an effect on the number of college grads. Another analysis he was asking was if there is a relationship between number of college grads and number of Jobs for each of the years.

To answer those questions, I used a Seaborn Heat Map. With this single Seaborn Heat Map, I was able to answer for both questions since it visualize the correlation between each variables very nicely. 

Fom the map I generated, I can conclude that neither LandArea nor NatAmenity have a high impact on the number of college grads. Still, there's a moderately high positive correlation between LandArea and college grad in 2000 since the correlation value is 0.73. Also, there's a pretty low correlation between NatAmenity and college grad in 1970 since the correlation value is 0.27. 

For the second question, I can conclude that the correlation between jobs and college grads increases as times go by from 1970 to 2000 in general. While the correlation is about 0.61-0.70 between jobs and college grads in 1970, the correlation is about 0.81-0.90 in 2000. Therefore, there's very high correlationship between number of jobs and number of college grads in 2000.
